**Q1.** This question is a case study for $k$ nearest neighbor The target variable `y` is `price` and the features are `year` and `mileage`.

1. Load the `./data/USA_cars_datasets.csv`. Keep the following variables and drop the rest: `price`, `year`, `mileage`. Are there any `NA`'s to handle? Look at the head and dimensions of the data.
2. Maxmin normalize `year` and `mileage`.
3. Split the sample into ~80% for training and ~20% for evaluation.
4. Use the $k$NN algorithm and the training data to predict `price` using `year` and `mileage` for the test set for $k=3,10,25,50,100,300$. For each value of $k$, compute the mean squared error and print a scatterplot showing the test value plotted against the predicted value. What patterns do you notice as you increase $k$?
5. Determine the optimal $k$ for these data.
6. Describe what happened in the plots of predicted versus actual prices as $k$ varied, taking your answer into part 6 into account. (Hint: Use the words "underfitting" and "overfitting".)

````python
# PART 1
import numpy as np
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/annikarichardson2024/KNN/main/USA_cars_datasets.csv')
df = df.loc[:,['price','year','mileage'] ]
print(df.shape)

# dimensions are 2499 observations and 3 variables
````

````python
# PART 1
df.head()
df.describe()

# there are not any NAs in the data. We know this because the observation count for each variable returns 2499
````

````python
# PART 2

#max min formula

def maxmin(x):
    mm = (x-min(x))/(max(x)-min(x))
    return mm

df['year'] = maxmin(df['year'])  # normalize year

df['mileage'] = maxmin(df['mileage']) # normalize mileage
````

````python
# PART 3

from sklearn.model_selection import train_test_split

y = df['price']
x = df.drop('price',axis=1)

# training 80%, testing 20%
trainX, testX, trainY, testY = train_test_split(x,y, # feature and target variables
                                                test_size=.2, # split sample 80 train and 20 test
                                                random_state=100) # for replication
````

````python
# PART 4


from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt

for k in [3,10,25,50,100,300]:
    model = KNeighborsRegressor(n_neighbors=k).fit(trainX,trainY) # Create a sk model for k
    y_hat = model.predict(testX) # Predict values for test set
    SSE = np.sum( (testY-y_hat)**2 ) # Save the computed SSE for test set

    plot, axes = plt.subplots()
    plt.scatter(testY,y_hat) # test value against predicted value
    plt.title('k: '+str(k)+', SSE: '+str(SSE))
    axes.set_ylim(-2000, 60000)
    axes.set_xlim(-2000, 60000)
    plt.show()

    # as K increases, so does the spread of the data points. They become more densely populated
    # toward 0 and less spread along the Y axis as the K value rises.
````

````python
# PART 5

from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt

kbar = 50  # Number of k's to try
kgrid = np.arange(1,kbar) # range of ks
SSE = np.zeros(kbar)  # store the SSE

for k in range(kbar):
    #fitted_model = KNeighborsRegressor(n_neighbors=k+1).fit(trainX,trainY) # create sk model for k and train the model on data
    model = KNeighborsRegressor(n_neighbors=k+1) # Create a sk model for k
    fitted_model = model.fit(trainX,trainY) # Train the model on our data
    y_hat = fitted_model.predict(testX) # predicted values for test set
    SSE[k] = np.sum( (testY-y_hat)**2 ) # save  computed SSE
 # optimal k above



SSE_min = np.min(SSE) # Lowest  SSE
index = np.where(SSE==SSE_min)  # indices of SSE that equal the minimum
kstar = kgrid[index] # optimal k


plt.plot(np.arange(0,kbar),SSE) # Plot SSE on y axis by k
plt.show()

print("optimal k:"+str(kstar)) # optimal is 48
````

````python
# PART 6

# the optimal k is 77. So based on that knowledge and the plot above, we can determine that
# the model likely overfits for k=300, and undertirs for 3, 10, and 25. For 50 and 100, the situation likely varies.
# An increase in K leads to a decrease in the range of predictions because outliers start to be less impactful on the average.
#
````
